# Pivot Table

```python
import json
import pandas as pd
import numpy as np

with open("transaction-data-adhoc-analysis.json","r") as f:
    tdata = json.load(f)

df = pd.DataFrame(tdata)
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['month'] = df['transaction_date'].dt.month_name()
```

```python
df
```

```python
import csv

df.to_csv(r'/Users/dianneprado/github-repo/project-adhoc/data.csv', index=False)
```

### __Total sales per month of each customer__

```python
index_input = ["name","username"]
columns_input = ["month"]  
df1 = df.pivot_table(index=index_input, columns=columns_input, aggfunc={'transaction_value': 'sum'},fill_value=0)
df1['transaction_value']
```

```python
import csv

df1.to_csv(r'/Users/dianneprado/github-repo/project-adhoc/tsm-data.csv', index=True, header=True)
```

### __Count of customer purchase per month__

```python
df2 = df.groupby(['name', 'month']).size().unstack(fill_value=0)
df2
```

```python
import csv

df2.to_csv(r'/Users/dianneprado/github-repo/project-adhoc/cpm-data.csv', index=True)
```
